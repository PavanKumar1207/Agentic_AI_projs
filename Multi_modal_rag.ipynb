```python
from google.generativeai import GenerativeModel
from google.generativeai import types
import os

def image_summarize_gemini_clean(image_path):
    """Generate a clean, concise summary for an image using Gemini"""
    with open(image_path, "rb") as f:
        image_bytes = f.read()

    model = GenerativeModel('gemini-2.0-flash')
    response = model.generate_content(
    contents=[
        {"mime_type": "image/jpeg", "data": image_bytes},
        "Provide a single, concise, one-sentence summary of this image optimized for retrieval."
    ]
)

    return response.text.strip()
```

```python
def generate_img_summaries(path):
    """
    Generate clean image summaries using Gemini for all .jpg images in the given path.
    Returns:
        - List of image file names
        - List of clean summaries
    """
    image_files = []
    image_summaries = []

    for img_file in sorted(os.listdir(path)):
        if img_file.lower().endswith(".jpg"):
            img_path = os.path.join(path, img_file)
            image_files.append(img_file)
            summary = image_summarize_gemini_clean(img_path)
            image_summaries.append(summary)

    return image_files, image_summaries
```

```python
import uuid
from langchain.schema import Document
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.retrievers.multi_vector import MultiVectorRetriever
from langchain.storage import InMemoryStore

def create_multi_vector_retriever(vectorstore, text_summaries, texts, table_summaries, tables, image_summaries, images):
    """
    Create retriever that indexes summaries, but returns raw images or texts
    """

    # Assuming InMemoryStore and MultiVectorRetriever are imported or defined elsewhere
    store = InMemoryStore()
    id_key = "doc_id"

    retriever = MultiVectorRetriever(
        vectorstore=vectorstore,
        docstore=store,
        id_key=id_key,
    )

    def add_documents(retriever, doc_summaries, doc_contents):
        doc_ids = [str(uuid.uuid4()) for _ in doc_contents]

        summary_docs = [
            Document(page_content=s, metadata={id_key: doc_ids[i]})
            for i, s in enumerate(doc_summaries)
        ]

        retriever.vectorstore.add_documents(summary_docs)
        retriever.docstore.mset(list(zip(doc_ids, doc_contents)))

    if text_summaries:
        add_documents(retriever, text_summaries, texts)
    if table_summaries:
        add_documents(retriever, table_summaries, tables)
    if image_summaries:
        add_documents(retriever, image_summaries, images)

    return retriever

# Use HuggingFaceEmbeddings with a popular sentence-transformer model
embedding_function = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

vectorstore = Chroma(
    collection_name="mm_rag",
    embedding_function=embedding_function,
    persist_directory="./chroma_db"
)

# Example usage:
retriever_multi_vector_img = create_multi_vector_retriever(
    vectorstore,
    text_summaries,  # list of text summary strings
    Text,            # list of raw text documents
    table_summaries, # list of table summary strings
    tab,           # list of raw tables
    image_summaries, # list of image summary strings
    image_files, # list of images encoded in base64 or raw format
)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.schema.runnable import RunnableLambda, RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain_core.messages import HumanMessage # Ensure this is imported if not already

def multi_modal_rag_chain(retriever):
    model = ChatGoogleGenerativeAI(model="gemini-pro-vision", temperature=0, max_tokens=1024)

    chain = (
        {
            "context": retriever | RunnableLambda(split_image_text_types),
            "question": RunnablePassthrough(),
        }
        | RunnableLambda(img_prompt_func)  # Returns {"messages": [...]}
        | model
        | StrOutputParser()
    )

    return chain
```

## Without displaying retrieved Images in output

```python
from PIL import Image
import base64
from io import BytesIO
import os
from langchain.schema import Document

IMAGE_FOLDER = "/content/extracted_data/"

# 🔁 Converts image to base64 with data URI scheme
def encode_image_to_base64(image: Image.Image) -> str:
    buffered = BytesIO()
    image.save(buffered, format="JPEG")  # or "PNG" depending on your image
    img_str = base64.b64encode(buffered.getvalue()).decode("utf-8")
    return f"data:image/jpeg;base64,{img_str}"


# 🔁 Separates image and text, loads images and encodes them
def split_image_text_types(docs):
    texts = []
    images = []

    for doc in docs:
        content = doc.page_content if hasattr(doc, "page_content") else doc

        if isinstance(content, str) and content.lower().endswith((".jpg", ".jpeg", ".png")):
            img_path = os.path.join(IMAGE_FOLDER, content)
            try:
                img = Image.open(img_path)
                encoded_img = encode_image_to_base64(img)
                images.append(encoded_img)
            except Exception as e:
                print(f"Error loading image {img_path}: {e}")
        else:
            texts.append(content)

    return {"texts": texts, "images": images}


# 🧠 Builds the multimodal prompt with base64 images
def img_prompt_func(inputs):
    context = inputs["context"]
    question = inputs["question"]

    texts = context.get("texts", [])
    images = context.get("images", [])

    context_text = "\n".join(texts)

    content = [{"type": "text", "text": f"Context:\n{context_text}\n\nQuestion: {question}"}]

    for base64_img in images:
        content.append({
            "type": "image_url",
            "image_url": {
                "url": base64_img
            }
        })

    return [
        {
            "role": "user",
            "content": content
        }
    ]


# 🔗 RAG Chain construction
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.schema.runnable import RunnableLambda, RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

def multi_modal_rag_chain(retriever):
    model = ChatGoogleGenerativeAI(model="gemini-2.0-flash-001", temperature=0, max_tokens=1024)

    chain = (
        {
            "context": retriever | RunnableLambda(split_image_text_types),
            "question": RunnablePassthrough(),
        }
        | RunnableLambda(img_prompt_func)
        | model
        | StrOutputParser()
    )

    return chain
```

```python
query = "What are robotic milestones with years?"

chain = multi_modal_rag_chain(retriever_multi_vector_img)
response = chain.invoke(query)
```
